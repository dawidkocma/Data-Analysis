```python
# College major with the lowest starting salary
print(clean_df['Starting Median Salary'].min()) #index 49
print(clean_df['Undergraduate Major'].loc[clean_df['Starting Median Salary'].idxmin()])
```

```python
# How much do graduates earn after university
clean_df[['Undergraduate Major','Starting Median Salary']]
```

```python
# Which university major has the lowest mid-career salary
major_idx = clean_df['Mid-Career Median Salary'].idxmin()
clean_df.loc[major_idx][['Undergraduate Major', 'Mid-Career Median Salary']]
```

```python
# Find Lowest Risk Majors
# lowest risk majors are the ones where there is a small difference between the lowest and the highest salaries. 

spread_col = clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])

# Insert new dataframe at position 1
# Check if 'Spread' column exists in the DF.
if 'Spread' not in clean_df.columns:
    # Insert 'Spread' column at the second position (index 1)
    clean_df.insert(1, 'Spread', spread_col)
    
# Display the fist few rows of the DF        
clean_df.head()

# Sort by the lowest spread
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']]
```

```python
# Find the degrees with the highest pay and the lowest spread. This will show the highest potential degrees
high_potential = clean_df.sort_values(by=['Mid-Career 90th Percentile Salary', "Spread"], ascending=[False, True])

# Display the first 5 rows
high_potential.head()
```

```python
# Calculate number of University majors for each group
clean_df.groupby('Group').count()
```

```python
# Find mean salary for each group
# .mean() cannot be applied to non-numeric columns so you need to select numeric only parameter
clean_df.groupby('Group').mean(numeric_only=True)
```

```python
# Format the output
pd.options.display.float_format = '{:,.2f}'.format
clean_df.groupby('Group').mean(numeric_only=True)
```
